```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from wordcloud import WordCloud
```

```python
df=pd.read_csv('/content/drive/MyDrive/tweets_with_predictions.csv')
```

```python
df['polarity']= df['polarity'].replace({1:'positive',0:'negative'})
df
```

```python
#Pie chart
sentiment_counts = df.groupby('polarity').count()
sizes = sentiment_counts['clean_text']
plt.pie(sizes, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Sentiment Analysis')
plt.show()
```

```python
# Group the data by polarity
polarity_groups = df.groupby('polarity')

# Create a word cloud for each polarity
for polarity, group in polarity_groups:
    text = ' '.join(group['clean_text'])
    wordcloud = WordCloud(max_font_size=50, max_words=100, background_color='white').generate(text)
    plt.figure()
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(polarity)
plt.show()
```

```python
grouped_data = df.groupby('polarity')['clean_text'].count().reset_index()

# create the bar chart using seaborn
sns.barplot(x='polarity', y='clean_text', data=grouped_data)

# set the title and labels
plt.title('Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Count')

# show the chart
plt.show()
```
